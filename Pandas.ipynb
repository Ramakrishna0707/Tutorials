```python
# Renaming the LanguageWorkedWith column to Python 
python_df.rename(columns= {'LanguageWorkedWith': 'NumKnowsPython','Country': 
                           'NumResponds'},inplace = True)
```

```python
python_df
```

```python
python_df['PctKnowsPython'] = (python_df['NumKnowsPython']/python_df['NumResponds']) * 100
```
